```python
dept_a_purchases = pd.read_excel("dept_A_purchases.xlsx")
dept_b_purchases = pd.read_excel("dept_B_purchases.xlsx")
dept_c_purchases = pd.read_excel("dept_C_purchases.xlsx")

dept_c_purchases.head()
```
